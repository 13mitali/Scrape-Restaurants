```
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys

# Configure Chrome options for Google Colab
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run in headless mode (no GUI)
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Initialize WebDriver
service = Service('/usr/bin/chromedriver')
browser = webdriver.Chrome(service=service, options=options)

# Google Maps search URL for restaurants in Pune
search_url = "https://www.google.com/maps/search/restaurants+in+Pune"

# Open search URL
browser.get(search_url)
time.sleep(5)  # Wait for page to load

# Scroll to load more results
# Instead of relying on a specific class, use a more general selector
# that targets the scrollable element based on its role or attributes
scrollable_div = WebDriverWait(browser, 10).until(
    EC.presence_of_element_located((By.XPATH, "//div[contains(@aria-label, 'Results for')]"))
)

# or
# scrollable_div = browser.find_element(By.CSS_SELECTOR, '[role="feed"]')

for _ in range(20):  # Scroll 20 times to load more restaurants
    browser.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scrollable_div)
    time.sleep(2)  # Give time for new results to load


# Extract restaurant links
restaurant_links = []
try:
    restaurant_elements = WebDriverWait(browser, 15).until(
        EC.presence_of_all_elements_located((By.XPATH, "//a[contains(@href, '/maps/place/')]"))
    )
    restaurant_links = list(set([elem.get_attribute("href") for elem in restaurant_elements]))  # Remove duplicates
    restaurant_links = restaurant_links[:50]  # Get first 50 links
except:
    print("Failed to fetch restaurant links.")

# List to store restaurant data
restaurant_data = []

# Loop through each restaurant URL
for i, url in enumerate(restaurant_links):
    browser.get(url)
    time.sleep(3)

    try:
        name = WebDriverWait(browser, 10).until(
            EC.presence_of_element_located((By.XPATH, "//h1[contains(@class, 'DUwDvf')]"))
        ).text
    except:
        name = None

    try:
        rating = WebDriverWait(browser, 10).until(
            EC.presence_of_element_located((By.XPATH, "//span[@class='MW4etd']"))
        ).text
    except:
        rating = None

    try:
        phone = WebDriverWait(browser, 10).until(
            EC.presence_of_element_located((By.XPATH, "//button[contains(@data-tooltip, 'Copy phone number')]"))
        ).text
    except:
        phone = None

    try:
        address = WebDriverWait(browser, 10).until(
            EC.presence_of_element_located((By.XPATH, "//button[contains(@data-tooltip, 'Copy address')]"))
        ).text
    except:
        address = None

    # Store data in list
    restaurant_data.append([name, rating, phone, address])
    print(f"{i+1}. Scraped: {name}")

# Close browser
browser.quit()

# Create a Pandas DataFrame
df = pd.DataFrame(restaurant_data, columns=["Restaurant Name", "Rating", "Phone Number", "Address"])

# Handle missing values
df.fillna("Not Available", inplace=True)

# Save to CSV file
csv_filename = "restaurants_pune.csv"
df.to_csv(csv_filename, index=False)

print(f"\n✅ Data saved to {csv_filename}")

# Display the DataFrame
df
```
